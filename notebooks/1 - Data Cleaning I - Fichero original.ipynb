```python
# TODO!!!!! Cambiar a read_csv indicando el número de filas a cargar y guardar en fichero aparte para subir a modo de prueba
# Lectura de las primeras 20 filas para ver que tipos de dato contiene cada columna
rows_array = []
columns_array = []

with open(SOURCE_DATA_FILENAME, "r") as open_file:
    array_line = []
    currentLine = 0;
    for line in open_file:
        array_line = line.split("|");
        if currentLine == 0:
            columns_array = array_line
        else:
            rows_array.append(array_line)
            
        currentLine += 1;        
        if currentLine >= 20:
            break;
    
open_file.close()

df = pd.DataFrame(rows_array, columns=columns_array)
df
```
